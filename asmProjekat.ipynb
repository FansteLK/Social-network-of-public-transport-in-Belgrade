```python
import datetime as time
from operator import mul



def timeToMinutes(x):
  factors = (60, 1, 1/60)
  x = sum(i*j for i, j in zip(map(int, x.split(':')), factors))
  return x
```

```python
#%pip install python-docx
import docx
import pandas as pd
#Ispis df-ova u dockx
# i am not sure how you are getting your data, but you said it is a
# pandas data frame
def df_toDocx(output,data):
    


    # open an existing document
    doc = docx.Document()

    # add a table to the end and create a reference variable
    # extra row is so we can add the header row
    t = doc.add_table(data.shape[0]+1, data.shape[1])

    # add the header rows.
    for j in range(data.shape[-1]):
        t.cell(0,j).text = data.columns[j]

# add the rest of the data frame
    for i in range(data.shape[0]):
        for j in range(data.shape[-1]):
            t.cell(i+1,j).text = str(data.values[i,j])

# save the doc
    print(doc)
    doc.save(output)
```

```python
# 2.1
pd.set_option('display.max_rows', None)
pd.set_option('display.min_rows', 50)
#spajanje dataseta stoptimes i trips
import numpy as np
import functools
data=dataStopTimes.merge(dataTrips)
#print(data)
data['number']=0
# Odredjivanje krajnjeg i pocetnog polaska a zatim i broj polasaka kao i prosecno vreme izmedju 2 za svaki od smerova za sve rute ->result
result = data.groupby(['route_id','service_id','direction_id']).agg(
    number=('route_id','size'),
    Min=('arrival_time', 'min'),
    Max=('arrival_time', 'max'))
result.columns=['number','Min','Max']
result['Min']=result['Min'].apply(timeToMinutes)
result['Max']=result['Max'].apply(timeToMinutes)
result['diff']=result['Max']-result['Min']
result['polasci']=result['diff']/result['number']

if 'Min' in result.columns:
    result = result.drop(['Min'], axis=1)
if 'Max' in result.columns:
    result = result.drop(['Max'], axis=1)
if 'diff' in result.columns:
    result= result.drop(['diff'],axis=1)


result=result.reset_index();

#print (result['service_id'])
if 'number' in data.columns:
    data = data.drop(['number'], axis=1)



#Odredjivanje prosecnog vremena polaska usredjnjavanjem po smerovima ->result2

result2=result.groupby(['route_id','service_id']).agg(AVG=('polasci','mean'))

#Izbacivanje ruta sa samo jednim polaskom
mask=result2['AVG']>0
result2=result2[mask]
result2=result2.reset_index();

#Odredjivanje dana kojim odredjene rute operisu
result3 = result2.groupby(['route_id'])['service_id'].apply(','.join).reset_index()





#data=data.merge(result2)
#dodavanje ruta u odrejne skupove u odnosu na to kojim danima operisu
setSva3 = set()
setVikend=set()
setRSubota=set()
setRNedelja=set()
setRadni=set()
setSubota=set()
setNedelja=set()
for i in range(0 ,result3.shape[0]):
    if(('Zimski-Radni Dan' in result3.iloc[i]['service_id']) and ('Zimski-Subota' in result3.iloc[i]['service_id']) and 'Zimski-Nedelja' in result3.iloc[i]['service_id']):
        setSva3.add(result3.iloc[i]['route_id'])
    elif('Zimski-Subota' in result3.iloc[i]['service_id'] and 'Zimski-Nedelja' in result3.iloc[i]['service_id']):
        setVikend.add(result3.iloc[i]['route_id'])
    elif ('Zimski-Radni Dan' in result3.iloc[i]['service_id'] and 'Zimski-Subota' in result3.iloc[i]['service_id']):
        setRSubota.add(result3.iloc[i]['route_id'])
    elif('Zimski-Radni Dan' in result3.iloc[i]['service_id'] and 'Zimski-Nedelja' in result3.iloc[i]['service_id']):
        setRNedelja.add(result3.iloc[i]['route_id'])
    elif("Zimski-Radni Dan" in result3.iloc[i]['service_id']):
        setRadni.add(result3.iloc[i]['route_id'])
    elif ('Zimski-Subota' in result3.iloc[i]['service_id']):
        setSubota.add(result3.iloc[i]['route_id'])
    elif ('Zimski-Nedelja' in result3.iloc[i]['service_id']):
        setNedelja.add(result3.iloc[i]['route_id'])

#print(len(setSva3))
#print(len(setVikend))
#print(len(setRSubota))
#print(len(setRNedelja))
#print(len(setRadni))
#print(len(setSubota))
#print(len(setNedelja))



#izracunavanje prosecnog vremena izmedju polazaka u odnosu na kategoriju gde rute upadaju
for i in setSva3:
    mask=((result2['service_id'] =='Zimski-Radni Dan') & (result2['route_id']==i))
    result2.update(result2[mask]['AVG'].mul(5))
    
for i in setRSubota:
    mask=((result2['service_id'] =='Zimski-Radni Dan') & (result2['route_id']==i))
    mask2=((result2['service_id'] =='Zimski-Subota') & (result2['route_id']==i))
    result2.update(result2[mask]['AVG'].mul(35/6))
    result2.update(result2[mask2]['AVG'].mul(7/6))
for i in setRadni:
     mask=((result2['service_id'] =='Zimski-Radni Dan') & (result2['route_id']==i))
     result2.update(result2[mask]['AVG'].mul(7))
result4= result2.groupby(['route_id'],as_index=False).agg(SUM=('AVG','sum'))
result4.update(result4['SUM'].div(7))
result4=result4.sort_values(by='SUM')
#df.sort_values(by='col1', ascending=False)

result5=result4.merge(dataDayRoutes)[['route_id','route_short_name','SUM']]

display(result5)
df_toDocx('./docks/prosecno_vreme_izmedju_polazaka.docx',result5)

```
